```python
import numpy as np
import matplotlib.pyplot as plt

def true_function(x):
    return np.sin(np.pi * x * 0.8) * 10

# xの値を生成する
x_values = np.linspace(-1, 1, 1000)  # -1から1までの間を1000点で均等に分割する

# 関数の呼び出しと結果の取得
y_values = true_function(x_values)

# グラフの描画
plt.plot(x_values, y_values)
plt.title('True Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
